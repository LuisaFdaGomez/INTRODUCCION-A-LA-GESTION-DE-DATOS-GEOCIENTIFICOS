##Proyecciones

### Ejercicio 1
Cambiar la proyección de la capa Depto_3 al actual sistema de referencia con origen nacional


```python
## A continuación se presenta una explicación para realizar el cambio de proyeeción de una capa vector

import geopandas as gpd

# Cargar el shapefile original
gdf_original = gpd.read_file('ruta/al/archivo.shp')

# Definir la proyección objetivo (por ejemplo, EPSG:4326 para WGS84)
proyeccion_objetivo = 'EPSG:4326'

# Cambiar la proyección del GeoDataFrame
gdf_nuevo = gdf_original.to_crs(proyeccion_objetivo)

# Guardar el nuevo GeoDataFrame en un nuevo archivo shapefile
gdf_nuevo.to_file('ruta/al/nuevo_archivo.shp')
```


Ejercicio 2 :Cambiar la proyección de la imgen del aeropuerto al actual sistema de referencia con origen nacional

```python
## cambio de proyección de una capa raster

from osgeo import gdal

# Ruta al archivo raster original
ruta_original = 'ruta/a/tu/raster/original.tif'

# Ruta al nuevo archivo raster con el sistema de referencia cambiado
ruta_nueva = 'ruta/a/tu/raster/nuevo.tif'

# Definir el nuevo sistema de referencia (por ejemplo, EPSG:4326 para WGS84)
nuevo_srs = 'EPSG:4326'

# Abrir el archivo raster original
raster_original = gdal.Open(ruta_original)

# Reproyectar la imagen raster al nuevo sistema de referencia
raster_nuevo = gdal.Warp(ruta_nueva, raster_original, dstSRS=nuevo_srs)

# Cerrar el archivo original y el nuevo archivo raster
raster_original = None
raster_nuevo = None
```
