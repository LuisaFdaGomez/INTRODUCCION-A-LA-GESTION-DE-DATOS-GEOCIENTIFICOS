```python
import gpxpy

# Ruta al archivo GPX
ruta_al_archivo_gpx = '/content/drive/MyDrive/Gestion_Datos/11303903.gpx'

# Lee el archivo GPX
with open(ruta_al_archivo_gpx, 'r') as gpx_file:
    gpx = gpxpy.parse(gpx_file)
```

```python
# Acceder a los tracks del archivo GPX
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            print('Track Point at ({0},{1}) -> {2}'.format(point.latitude, point.longitude, point.elevation))

# Acceder a los waypoints del archivo GPX
for waypoint in gpx.waypoints:
    print('Waypoint at ({0},{1}) -> {2}'.format(waypoint.latitude, waypoint.longitude, waypoint.name))

# Acceder a las rutas del archivo GPX
for route in gpx.routes:
    print('Route:')
    for point in route.points:
        print('Point at ({0},{1}) -> {2}'.format(point.latitude, point.longitude, point.elevation))
```
